### variables

```python
total = {}


total["poblacion"] = []
total["Edad0-15"] = []
total["Edad16-30"] = []
total["Edad31-45"] = []
total["Edad46-60"] = []
total["Edad61-75"] = []
total["Edad76-90"] = []
total["Edad91-105"] = []
total["hombres"] = []
total["mujeres"] = []
total["intubado"] = []
#total["no_especificado"] = []



total["defunciones"] = []
#total["recuperados"] = []
#total["activos"] = []
total["hospitalizados"] = []
```

```python
d.head(10)
```

```python
d['edad'] = pd.to_numeric(d['edad'], errors='coerce')
d.fallecido.value_counts()
d['fallecido'] = d['fallecido'].map({"SI": 1, "NO": 0})
d['fecha_internacion'] = d['fecha_internacion'].astype(str)
d.cuidado_intensivo = d.cuidado_intensivo.map({"SI": 1, "NO": 0})
d.asistencia_respiratoria_mecanica = d.asistencia_respiratoria_mecanica.map({"SI": 1, "NO": 0})
```

```python
d['fecha_internacion'].value_counts()
```

### f

```python
bins = [0, 15, 30, 45, 60, 75, 90, 105]
labels = [
        "Edad0-15",
        "Edad16-30",
        "Edad31-45",
        "Edad46-60",
        "Edad61-75",
        "Edad76-90",
        "Edad91-105",
]
age_groups = pd.cut(d.edad, bins=bins, labels=labels, right=False).value_counts()
for label in labels:
        total[label].append(age_groups.get(label, 0))
total["poblacion"].append(d.shape[0])
total["defunciones"].append((d.fallecido==1).sum())
sexo_counts = d.sexo.value_counts()
total["hombres"].append(sexo_counts.get("M", 0))
total["mujeres"].append(sexo_counts.get("F", 0))
#tipo_paciente_counts = d.tipo_paciente.value_counts()
total["hospitalizados"].append((d.fecha_internacion!="None").sum())

asi_re = d.asistencia_respiratoria_mecanica.value_counts()[1]
total["intubado"].append(asi_re)

#total["ambulatorios"].append(tipo_paciente_counts.get(1, 0))
```

```python
total
totales = pd.DataFrame.from_dict(total)
totales.to_csv("argentina2020-2021.csv")
```
